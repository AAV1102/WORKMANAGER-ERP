````python

# PASO 2: Crear funciones de mapeo y normalización de datos

def normalizar_texto(texto):
    """Normaliza texto: elimina espacios extras, convierte a mayúsculas"""
    if pd.isna(texto):
        return ''
    return str(texto).strip().upper()

def mapear_columna(df, posibles_nombres):
    """
    Busca una columna en el DataFrame usando múltiples nombres posibles
    Retorna el nombre de la columna encontrada o None
    """
    for nombre in posibles_nombres:
        # Buscar coincidencia exacta (case insensitive)
        for col in df.columns:
            if str(col).strip().upper() == nombre.upper():
                return col
        # Buscar coincidencia parcial
        for col in df.columns:
            if nombre.upper() in str(col).strip().upper():
                return col
    return None

# Diccionario de mapeo: columna destino -> posibles nombres en archivos fuente
mapeo_columnas = {
    'SERIAL': ['SERIAL', 'SERIE', 'SERIE No.', 'SERIE NO', 'SERIAL No.', 'NUMERO DE SERIE'],
    'PLACA': ['PLACA', 'PLACA No.', 'PLACA NO', 'PLACA NUM', 'NUMERO PLACA'],
    'SEDE': ['SEDE', 'CIUDAD', 'UBICACION', 'LOCATION', 'SITE'],
    'TECNOLOGIA': ['TECNOLOGIA', 'TIPO', 'DESCRIPCION', 'DESCRIPCION DE LOS ACTIVOS', 'TIPO DE EQUIPO'],
    'MARCA': ['MARCA', 'BRAND', 'FABRICANTE'],
    'MODELO': ['MODELO', 'MODEL'],
    'PROCESADOR': ['PROCESADOR', 'PROCESSOR', 'CPU'],
    'CANTIDAD RAM': ['CANTIDAD RAM', 'RAM', 'MEMORIA', 'MEMORIA RAM'],
    'ARQUITECTURA RAM': ['ARQUITECTURA RAM', 'ARQUI RAM', 'TIPO RAM'],
    'TIPO DE DISCO': ['TIPO DE DISCO', 'TIPO DISCO', 'DISK TYPE'],
    'ALMACENAMIENTO': ['ALMACENAMIENTO', 'ESPACIO DISCO', 'DISCO', 'STORAGE', 'CAPACIDAD'],
    'SO': ['SO', 'SISTEMA OPERATIVO', 'OS', 'OPERATING SYSTEM'],
    'ESTADO': ['ESTADO', 'ESTADO ACTUAL', 'STATUS', 'CONDICION'],
    'ASIGNADO A': ['ASIGNADO A', 'ASIGNADO NUEVO', 'USUARIO', 'RESPONSABLE', 'ASIGNADO'],
    'ANTERIOR ASIGNADO': ['ANTERIOR ASIGNADO', 'USUARIO ANTERIOR', 'ASIGNADO ANTERIOR'],
    'AREA': ['AREA', 'DEPARTAMENTO', 'DEPT'],
    'FECHA ASIGNACION': ['FECHA ASIGNACION', 'FECHA', 'FECHA DE ASIGNACION'],
    'FECHA DE LLEGADA': ['FECHA DE LLEGADA', 'FECHA LLEGADA', 'FECHA INGRESO'],
    'OBSERVACIONES': ['OBSERVACIONES', 'OBSERVACION', 'NOTAS', 'COMENTARIOS'],
    'DISPONIBLE': ['DISPONIBLE', 'DISPONIBILIDAD', 'AVAILABLE'],
    'ENTRADA OC COMPRA': ['ENTRADA OC COMPRA', 'OC COMPRA', 'ORDEN COMPRA'],
    'OC': ['OC', 'ORDEN', 'ORDEN DE COMPRA'],
    'PROVEEDOR': ['PROVEEDOR', 'SUPPLIER', 'VENDOR'],
    'SERIAL MONITOR': ['SERIAL MONITOR', 'SERIE MONITOR', 'SERIAL2'],
    'PLACA MONITOR': ['PLACA MONITOR', 'PLACA DE MONITOR'],
    'MARCA DE MONITOR': ['MARCA DE MONITOR', 'MARCA MONITOR'],
    'MODELO MONITOR': ['MODELO MONITOR', 'MODELO DE MONITOR'],
    'ENVIADO NIT': ['ENVIADO NIT', 'ENVIADO /CARGADO', 'ENVIADO', 'CARGADO']
}

print("✓ Funciones de mapeo y normalización creadas")
print(f"✓ {len(mapeo_columnas)} columnas con mapeo definido")
````

````python

# PASO 3: Función para generar códigos automáticos

def generar_codigo_sede(sede):
    """Genera código de 3 letras para la sede"""
    if pd.isna(sede) or sede == '':
        return 'GEN'
    
    sede_upper = str(sede).strip().upper()
    
    # Mapeo de sedes conocidas
    mapeo_sedes = {
        'MEDELLIN': 'MED',
        'CARTAGENA': 'CTG',
        'PASTO': 'PAS',
        'BOGOTA': 'BOG',
        'CALI': 'CAL',
        'BARRANQUILLA': 'BAQ',
        'BUCARAMANGA': 'BUC',
        'PEREIRA': 'PER',
        'MANIZALES': 'MAN',
        'CUCUTA': 'CUC'
    }
    
    for ciudad, codigo in mapeo_sedes.items():
        if ciudad in sede_upper:
            return codigo
    
    # Si no está en el mapeo, tomar las 3 primeras letras
    return sede_upper[:3]

def generar_codigo_tecnologia(tecnologia):
    """Genera código abreviado para tipo de tecnología"""
    if pd.isna(tecnologia) or tecnologia == '':
        return 'GEN'
    
    tech_upper = str(tecnologia).strip().upper()
    
    # Mapeo de tecnologías
    mapeo_tech = {
        'PORTATIL': 'PORT',
        'LAPTOP': 'PORT',
        'NOTEBOOK': 'PORT',
        'PC ESCRITORIO': 'DESK',
        'DESKTOP': 'DESK',
        'TODO EN UNO': 'AIO',
        'ALL IN ONE': 'AIO',
        'MONITOR': 'MON',
        'PANTALLA': 'MON',
        'TECLADO': 'TEC',
        'KEYBOARD': 'TEC',
        'MOUSE': 'MOU',
        'RATON': 'MOU',
        'IMPRESORA': 'IMP',
        'PRINTER': 'IMP',
        'TELEFONO': 'TEL',
        'PHONE': 'TEL',
        'SWITCH': 'SWI',
        'ROUTER': 'ROU',
        'SERVIDOR': 'SRV',
        'SERVER': 'SRV',
        'TABLET': 'TAB',
        'CELULAR': 'CEL',
        'MOVIL': 'CEL',
        'CCTV': 'CAM',
        'CAMARA': 'CAM',
        'ESCANER': 'ESC',
        'SCANNER': 'ESC',
        'UPS': 'UPS',
        'BASE': 'BAS',
        'DIADEMA': 'DIA',
        'HEADSET': 'DIA'
    }
    
    for tech, codigo in mapeo_tech.items():
        if tech in tech_upper:
            return codigo
    
    # Si no está en el mapeo, tomar las 3-4 primeras letras
    return tech_upper[:4]

# Probar las funciones
print("=== PRUEBAS DE GENERACIÓN DE CÓDIGOS ===\n")
print("Códigos de Sede:")
for sede in ['MEDELLIN', 'Cartagena', 'pasto', 'BOGOTA', 'NUEVA CIUDAD']:
    print(f"  {sede:20} -> {generar_codigo_sede(sede)}")

print("\nCódigos de Tecnología:")
for tech in ['PORTATIL', 'Monitor', 'TODO EN UNO', 'Telefono', 'IMPRESORA', 'SWITCH']:
    print(f"  {tech:20} -> {generar_codigo_tecnologia(tech)}")
````

````python

# PASO 4: Procesar y unificar datos de todos los archivos

def procesar_archivo(df, nombre_archivo):
    """
    Procesa un DataFrame y lo mapea a la estructura del inventario general
    """
    print(f"\n--- Procesando: {nombre_archivo} ---")
    
    # Crear DataFrame con la estructura del inventario general
    df_procesado = pd.DataFrame(columns=columnas_inventario_general)
    
    # Mapear cada columna
    for col_destino, posibles_nombres in mapeo_columnas.items():
        col_fuente = mapear_columna(df, posibles_nombres)
        if col_fuente:
            print(f"  ✓ {col_destino:25} <- {col_fuente}")
            df_procesado[col_destino] = df[col_fuente].apply(normalizar_texto)
        else:
            df_procesado[col_destino] = ''
    
    # Rellenar columnas no mapeadas con valores vacíos
    for col in columnas_inventario_general:
        if col not in df_procesado.columns:
            df_procesado[col] = ''
    
    return df_procesado

# Procesar archivo de Cartagena
print("="*80)
print("PROCESANDO ARCHIVOS")
print("="*80)

df_cartagena_procesado = procesar_archivo(df_cartagena, "INVENTARIO CARTAGENA")
print(f"\n  Resultado: {len(df_cartagena_procesado)} registros procesados")
````

````python

# Procesar archivo de Pasto (necesita limpieza especial)
# Primero, identificar las filas de datos reales

df_pasto_raw = pd.read_excel("/workspace/uploaded_files/Inventario Pasto.xlsx", header=None)

# Buscar la fila de encabezados (contiene "ITEM", "DESCRIPCION", etc.)
header_row = None
for idx, row in df_pasto_raw.iterrows():
    if 'ITEM' in str(row.values) and 'DESCRIPCION' in str(row.values):
        header_row = idx
        break

if header_row is not None:
    # Extraer encabezados
    headers = df_pasto_raw.iloc[header_row].values
    # Limpiar encabezados
    headers_clean = []
    for h in headers:
        if pd.notna(h) and str(h).strip() != '':
            headers_clean.append(str(h).strip())
        else:
            headers_clean.append(f'Col_{len(headers_clean)}')
    
    # Crear DataFrame con datos
    df_pasto_data = df_pasto_raw.iloc[header_row+1:].copy()
    df_pasto_data.columns = headers_clean
    
    # Eliminar filas vacías
    df_pasto_data = df_pasto_data.dropna(how='all')
    
    # Filtrar solo filas con datos válidos (que tengan ITEM numérico o descripción)
    df_pasto_data = df_pasto_data[df_pasto_data['DESCRIPCION DE LOS ACTIVOS'].notna()]
    
    print(f"✓ Archivo Pasto procesado: {len(df_pasto_data)} registros")
    print(f"  Columnas: {df_pasto_data.columns.tolist()}")
    print(f"\n  Primeras 5 filas:")
    print(df_pasto_data[['DESCRIPCION DE LOS ACTIVOS', 'MARCA', 'PLACA No.', 'ESTADO ACTUAL', 'RESPONSABLE']].head())
    
    # Ahora procesar con la función de mapeo
    df_pasto_procesado = procesar_archivo(df_pasto_data, "INVENTARIO PASTO")
    print(f"\n  Resultado final: {len(df_pasto_procesado)} registros procesados")
else:
    print("✗ No se pudo encontrar la fila de encabezados en Pasto")
    df_pasto_procesado = pd.DataFrame(columns=columnas_inventario_general)
````

````python

# PASO 6: Generar códigos automáticos (ID, CODIGO UNIFICADO, PLACA individual)

import datetime

# Generar ID único secuencial
df_inventario_unificado['ID'] = range(1, len(df_inventario_unificado) + 1)

# Generar CODIGO INDIVIDUAL (PLACA si no existe)
# Formato: SEDE-TECH-###
codigos_individuales = []
contadores_sede_tech = {}

for idx, row in df_inventario_unificado.iterrows():
    sede = row['SEDE']
    tecnologia = row['TECNOLOGIA']
    placa_existente = row['PLACA']
    
    # Si ya tiene placa, usarla
    if placa_existente and placa_existente.strip() != '':
        codigos_individuales.append(placa_existente)
    else:
        # Generar código nuevo
        codigo_sede = generar_codigo_sede(sede)
        codigo_tech = generar_codigo_tecnologia(tecnologia)
        
        # Crear clave única para contador
        clave = f"{codigo_sede}_{codigo_tech}"
        
        if clave not in contadores_sede_tech:
            contadores_sede_tech[clave] = 1
        else:
            contadores_sede_tech[clave] += 1
        
        numero = str(contadores_sede_tech[clave]).zfill(3)
        codigo_individual = f"{codigo_sede}-{codigo_tech}-{numero}"
        codigos_individuales.append(codigo_individual)

# Actualizar PLACA con códigos individuales
df_inventario_unificado['PLACA'] = codigos_individuales

print(f"✓ Códigos individuales generados")
print(f"\nEjemplos de códigos generados:")
print(df_inventario_unificado[['SEDE', 'TECNOLOGIA', 'PLACA']].head(20))
````

````python

# PASO 7: Generar CODIGO UNIFICADO para equipos agrupados
# Por ahora, cada equipo tiene su propio código unificado
# En el futuro, se pueden agrupar equipos relacionados (ej: PC + Monitor + Mouse + Teclado)

# Para esta versión inicial, el código unificado será igual al individual
# Más adelante se puede implementar lógica de agrupación

df_inventario_unificado['CODIGO UNIFICADO'] = df_inventario_unificado['PLACA']

# Agregar metadatos de creación
df_inventario_unificado['CREADOR'] = 'SISTEMA AUTOMATICO'
df_inventario_unificado['FECHA CREACION'] = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
df_inventario_unificado['TRAZABILIDAD O MOVIMIENTOS'] = 'IMPORTACION INICIAL'
df_inventario_unificado['ACCIONES'] = ''

print("✓ Códigos unificados y metadatos agregados")
print(f"\nPrimeros registros con códigos:")
print(df_inventario_unificado[['ID', 'PLACA', 'CODIGO UNIFICADO', 'SEDE', 'TECNOLOGIA', 'ASIGNADO A']].head(10))
````

````python

# PASO 8: Normalizar estados para facilitar filtrado

def normalizar_estado(estado):
    """Normaliza los estados a categorías estándar"""
    if pd.isna(estado) or estado == '':
        return 'SIN DEFINIR'
    
    estado_upper = str(estado).strip().upper()
    
    # Mapeo de estados
    if 'BUEN' in estado_upper or 'ACTIVO' in estado_upper or 'NUEVO' in estado_upper:
        return 'ACTIVO'
    elif 'BAJA' in estado_upper or 'DADO DE BAJA' in estado_upper:
        return 'DADO DE BAJA'
    elif 'MAL' in estado_upper or 'DAÑADO' in estado_upper or 'ROTO' in estado_upper:
        return 'MAL ESTADO'
    elif 'REGULAR' in estado_upper:
        return 'REGULAR'
    elif 'PENDIENTE' in estado_upper or 'REVISION' in estado_upper:
        return 'PENDIENTE'
    elif 'PRESTAMO' in estado_upper:
        return 'PRESTAMO'
    elif 'REPOTENCIADO' in estado_upper:
        return 'REPOTENCIADO'
    elif 'TANDA' in estado_upper or 'NUEVO' in estado_upper:
        return 'TANDA NUEVA'
    else:
        return estado_upper

# Aplicar normalización
df_inventario_unificado['ESTADO_NORMALIZADO'] = df_inventario_unificado['ESTADO'].apply(normalizar_estado)

print("=== ESTADOS NORMALIZADOS ===")
print(df_inventario_unificado['ESTADO_NORMALIZADO'].value_counts())

# Mantener el estado original también
print("\n✓ Estados normalizados para facilitar filtrado")
````

````python

# PASO 9: Crear el archivo Excel con todas las pestañas

from openpyxl import Workbook
from openpyxl.styles import Font, Alignment, PatternFill, Border, Side
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.worksheet.table import Table, TableStyleInfo

# Crear libro de Excel
wb = Workbook()
wb.remove(wb.active)  # Eliminar hoja por defecto

# Estilos
header_fill = PatternFill(start_color="4472C4", end_color="4472C4", fill_type="solid")
header_font = Font(bold=True, color="FFFFFF", size=11)
center_align = Alignment(horizontal="center", vertical="center", wrap_text=True)
left_align = Alignment(horizontal="left", vertical="center", wrap_text=True)
thin_border = Border(
    left=Side(style="thin", color="000000"),
    right=Side(style="thin", color="000000"),
    top=Side(style="thin", color="000000"),
    bottom=Side(style="thin", color="000000")
)

def aplicar_estilo_tabla(ws, df):
    """Aplica estilo profesional a una hoja de Excel"""
    
    # Aplicar estilo a encabezados
    for cell in ws[1]:
        cell.fill = header_fill
        cell.font = header_font
        cell.alignment = center_align
        cell.border = thin_border
    
    # Aplicar bordes y alineación a datos
    for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=ws.max_column):
        for cell in row:
            cell.border = thin_border
            cell.alignment = left_align
    
    # Ajustar ancho de columnas
    for column in ws.columns:
        max_length = 0
        column_letter = column[0].column_letter
        for cell in column:
            try:
                if cell.value:
                    max_length = max(max_length, len(str(cell.value)))
            except:
                pass
        adjusted_width = min(max_length + 2, 50)  # Máximo 50 caracteres
        ws.column_dimensions[column_letter].width = adjusted_width
    
    # Congelar primera fila
    ws.freeze_panes = "A2"

print("Iniciando creación de archivo Excel con todas las pestañas...")
print("="*80)
````

````python

# CREAR SCRIPT DE IMPORTACIÓN UNIVERSAL
script_importador = '''
"""
IMPORTADOR UNIVERSAL DE INVENTARIOS
====================================
Este script permite importar datos de múltiples archivos Excel/CSV
y consolidarlos en el formato del Inventario General Unificado.

Uso:
    python importador_inventarios.py

Configuración:
    - Carpeta de origen: C:\\Users\\anderson.a\\Documents\\Inventario para importar
    - Archivo de salida: INVENTARIO_GENERAL_UNIFICADO.xlsx
"""

import pandas as pd
import os
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# CONFIGURACIÓN
# ============================================================================

CARPETA_ORIGEN = r"C:\\Users\\anderson.a\\Documents\\Inventario para importar"
ARCHIVO_SALIDA = "INVENTARIO_GENERAL_UNIFICADO.xlsx"

# Estructura completa del inventario general
COLUMNAS_INVENTARIO = [
    'ID', 'SERIAL', 'CODIGO UNIFICADO', 'PLACA', 'ANTERIOR PLACA', 'SEDE',
    'IP SEDE', 'TECNOLOGIA', 'MARCA', 'MODELO', 'MAC', 'SO', 'HOSTNAME', 'IP',
    'MARCA Y MODELO TELEMEDICINA', 'MOUSE', 'TECLADO', 'ANTERIOR ASIGNADO',
    'ASIGNADO A', 'CARGO', 'AREA', 'CONTACTO', 'FECHA ASIGNACION', 'CARGADO NIT',
    'ENVIADO NIT', 'FECHA ENVIADO', 'GUIA', 'PROCESADOR', 'ARQUITECTURA RAM',
    'CANTIDAD RAM', 'TIPO DE DISCO', 'ALMACENAMIENTO', 'PLACA MONITOR',
    'SERIAL MONITOR', 'MARCA DE MONITOR', 'MODELO MONITOR', 'ESTADO', 'DISPONIBLE',
    'OBSERVACIONES', 'SERIAL TELEMEDICINA', 'TIPO DE COMPONENTE ADICIONAL',
    'MARCA Y MODELO COMPONENTE ADICIONAL', 'SERIAL COMPONENTE ADICIONAL',
    'MARCA Y MODELO TELEFONO CORPORATIVO', 'SERIAL TELEFONO CORPORATIVO',
    'IMEI TELEFONO CORPORATIVO', 'MARCA Y MODELO IMPRESORA', 'IP IMPRESORA',
    'SERIAL IMPRESORA', 'PIN IMPRESORA', 'MARCA Y MODELO CCTV', 'SERIAL CCTV',
    'ENTRADA OC COMPRA', 'FECHA DE LLEGADA', 'OC', 'PROVEEDOR', 'MUEBLE ASIGNADO',
    'CREADOR', 'FECHA CREACION', 'TRAZABILIDAD O MOVIMIENTOS', 'ACCIONES'
]

# Mapeo de columnas (columna destino -> posibles nombres en archivos fuente)
MAPEO_COLUMNAS = {
    'SERIAL': ['SERIAL', 'SERIE', 'SERIE No.', 'SERIE NO', 'SERIAL No.', 'NUMERO DE SERIE'],
    'PLACA': ['PLACA', 'PLACA No.', 'PLACA NO', 'PLACA NUM', 'NUMERO PLACA'],
    'SEDE': ['SEDE', 'CIUDAD', 'UBICACION', 'LOCATION', 'SITE'],
    'TECNOLOGIA': ['TECNOLOGIA', 'TIPO', 'DESCRIPCION', 'DESCRIPCION DE LOS ACTIVOS', 'TIPO DE EQUIPO'],
    'MARCA': ['MARCA', 'BRAND', 'FABRICANTE'],
    'MODELO': ['MODELO', 'MODEL'],
    'PROCESADOR': ['PROCESADOR', 'PROCESSOR', 'CPU'],
    'CANTIDAD RAM': ['CANTIDAD RAM', 'RAM', 'MEMORIA', 'MEMORIA RAM'],
    'ARQUITECTURA RAM': ['ARQUITECTURA RAM', 'ARQUI RAM', 'TIPO RAM'],
    'TIPO DE DISCO': ['TIPO DE DISCO', 'TIPO DISCO', 'DISK TYPE'],
    'ALMACENAMIENTO': ['ALMACENAMIENTO', 'ESPACIO DISCO', 'DISCO', 'STORAGE', 'CAPACIDAD'],
    'SO': ['SO', 'SISTEMA OPERATIVO', 'OS', 'OPERATING SYSTEM'],
    'ESTADO': ['ESTADO', 'ESTADO ACTUAL', 'STATUS', 'CONDICION'],
    'ASIGNADO A': ['ASIGNADO A', 'ASIGNADO NUEVO', 'USUARIO', 'RESPONSABLE', 'ASIGNADO'],
    'ANTERIOR ASIGNADO': ['ANTERIOR ASIGNADO', 'USUARIO ANTERIOR', 'ASIGNADO ANTERIOR'],
    'AREA': ['AREA', 'DEPARTAMENTO', 'DEPT'],
    'FECHA ASIGNACION': ['FECHA ASIGNACION', 'FECHA', 'FECHA DE ASIGNACION'],
    'FECHA DE LLEGADA': ['FECHA DE LLEGADA', 'FECHA LLEGADA', 'FECHA INGRESO'],
    'OBSERVACIONES': ['OBSERVACIONES', 'OBSERVACION', 'NOTAS', 'COMENTARIOS'],
    'DISPONIBLE': ['DISPONIBLE', 'DISPONIBILIDAD', 'AVAILABLE'],
    'ENTRADA OC COMPRA': ['ENTRADA OC COMPRA', 'OC COMPRA', 'ORDEN COMPRA'],
    'OC': ['OC', 'ORDEN', 'ORDEN DE COMPRA'],
    'PROVEEDOR': ['PROVEEDOR', 'SUPPLIER', 'VENDOR'],
    'SERIAL MONITOR': ['SERIAL MONITOR', 'SERIE MONITOR', 'SERIAL2'],
    'PLACA MONITOR': ['PLACA MONITOR', 'PLACA DE MONITOR'],
    'MARCA DE MONITOR': ['MARCA DE MONITOR', 'MARCA MONITOR'],
    'MODELO MONITOR': ['MODELO MONITOR', 'MODELO DE MONITOR'],
    'ENVIADO NIT': ['ENVIADO NIT', 'ENVIADO /CARGADO', 'ENVIADO', 'CARGADO']
}

# ============================================================================
# FUNCIONES AUXILIARES
# ============================================================================

def normalizar_texto(texto):
    """Normaliza texto: elimina espacios extras, convierte a mayúsculas"""
    if pd.isna(texto):
        return ''
    return str(texto).strip().upper()

def mapear_columna(df, posibles_nombres):
    """Busca una columna en el DataFrame usando múltiples nombres posibles"""
    for nombre in posibles_nombres:
        # Buscar coincidencia exacta (case insensitive)
        for col in df.columns:
            if str(col).strip().upper() == nombre.upper():
                return col
        # Buscar coincidencia parcial
        for col in df.columns:
            if nombre.upper() in str(col).strip().upper():
                return col
    return None

def generar_codigo_sede(sede):
    """Genera código de 3 letras para la sede"""
    if pd.isna(sede) or sede == '':
        return 'GEN'
    
    sede_upper = str(sede).strip().upper()
    mapeo_sedes = {
        'MEDELLIN': 'MED', 'CARTAGENA': 'CTG', 'PASTO': 'PAS',
        'BOGOTA': 'BOG', 'CALI': 'CAL', 'BARRANQUILLA': 'BAQ',
        'BUCARAMANGA': 'BUC', 'PEREIRA': 'PER', 'MANIZALES': 'MAN', 'CUCUTA': 'CUC'
    }
    
    for ciudad, codigo in mapeo_sedes.items():
        if ciudad in sede_upper:
            return codigo
    
    return sede_upper[:3]

def generar_codigo_tecnologia(tecnologia):
    """Genera código abreviado para tipo de tecnología"""
    if pd.isna(tecnologia) or tecnologia == '':
        return 'GEN'
    
    tech_upper = str(tecnologia).strip().upper()
    mapeo_tech = {
        'PORTATIL': 'PORT', 'LAPTOP': 'PORT', 'NOTEBOOK': 'PORT',
        'PC ESCRITORIO': 'DESK', 'DESKTOP': 'DESK',
        'TODO EN UNO': 'AIO', 'ALL IN ONE': 'AIO',
        'MONITOR': 'MON', 'PANTALLA': 'MON',
        'TECLADO': 'TEC', 'KEYBOARD': 'TEC',
        'MOUSE': 'MOU', 'RATON': 'MOU',
        'IMPRESORA': 'IMP', 'PRINTER': 'IMP',
        'TELEFONO': 'TEL', 'PHONE': 'TEL',
        'SWITCH': 'SWI', 'ROUTER': 'ROU',
        'SERVIDOR': 'SRV', 'SERVER': 'SRV',
        'TABLET': 'TAB', 'CELULAR': 'CEL', 'MOVIL': 'CEL',
        'CCTV': 'CAM', 'CAMARA': 'CAM',
        'ESCANER': 'ESC', 'SCANNER': 'ESC',
        'UPS': 'UPS', 'BASE': 'BAS', 'DIADEMA': 'DIA', 'HEADSET': 'DIA'
    }
    
    for tech, codigo in mapeo_tech.items():
        if tech in tech_upper:
            return codigo
    
    return tech_upper[:4]

def procesar_archivo(filepath):
    """Procesa un archivo Excel o CSV y lo mapea a la estructura del inventario"""
    print(f"\\n  Procesando: {os.path.basename(filepath)}")
    
    try:
        # Intentar leer como Excel
        if filepath.endswith(('.xlsx', '.xls')):
            df = pd.read_excel(filepath)
        else:
            df = pd.read_csv(filepath, encoding='utf-8')
        
        # Crear DataFrame con la estructura del inventario general
        df_procesado = pd.DataFrame(columns=COLUMNAS_INVENTARIO)
        
        # Mapear cada columna
        columnas_mapeadas = 0
        for col_destino, posibles_nombres in MAPEO_COLUMNAS.items():
            col_fuente = mapear_columna(df, posibles_nombres)
            if col_fuente:
                df_procesado[col_destino] = df[col_fuente].apply(normalizar_texto)
                columnas_mapeadas += 1
        
        # Rellenar columnas no mapeadas con valores vacíos
        for col in COLUMNAS_INVENTARIO:
            if col not in df_procesado.columns or df_procesado[col].isna().all():
                df_procesado[col] = ''
        
        print(f"    ✓ {len(df_procesado)} registros, {columnas_mapeadas} columnas mapeadas")
        return df_procesado
        
    except Exception as e:
        print(f"    ✗ Error: {str(e)}")
        return pd.DataFrame(columns=COLUMNAS_INVENTARIO)

# ============================================================================
# PROCESO PRINCIPAL
# ============================================================================

def main():
    print("="*80)
    print("IMPORTADOR UNIVERSAL DE INVENTARIOS")
    print("="*80)
    
    # Verificar que existe la carpeta
    if not os.path.exists(CARPETA_ORIGEN):
        print(f"\\n✗ ERROR: No se encuentra la carpeta: {CARPETA_ORIGEN}")
        return
    
    # Buscar todos los archivos Excel y CSV
    archivos = []
    for file in os.listdir(CARPETA_ORIGEN):
        if file.endswith(('.xlsx', '.xls', '.csv')):
            archivos.append(os.path.join(CARPETA_ORIGEN, file))
    
    print(f"\\n✓ Encontrados {len(archivos)} archivos para procesar")
    
    # Procesar todos los archivos
    dfs_procesados = []
    for archivo in archivos:
        df_proc = procesar_archivo(archivo)
        if len(df_proc) > 0:
            dfs_procesados.append(df_proc)
    
    # Unificar todos los DataFrames
    if len(dfs_procesados) == 0:
        print("\\n✗ No se procesaron archivos exitosamente")
        return
    
    df_unificado = pd.concat(dfs_procesados, ignore_index=True)
    
    # Generar códigos automáticos
    print(f"\\n✓ Generando códigos automáticos...")
    df_unificado['ID'] = range(1, len(df_unificado) + 1)
    
    # Generar códigos individuales
    codigos_individuales = []
    contadores = {}
    
    for idx, row in df_unificado.iterrows():
        if row['PLACA'] and row['PLACA'].strip() != '':
            codigos_individuales.append(row['PLACA'])
        else:
            codigo_sede = generar_codigo_sede(row['SEDE'])
            codigo_tech = generar_codigo_tecnologia(row['TECNOLOGIA'])
            clave = f"{codigo_sede}_{codigo_tech}"
            
            contadores[clave] = contadores.get(clave, 0) + 1
            numero = str(contadores[clave]).zfill(3)
            codigo = f"{codigo_sede}-{codigo_tech}-{numero}"
            codigos_individuales.append(codigo)
    
    df_unificado['PLACA'] = codigos_individuales
    df_unificado['CODIGO UNIFICADO'] = codigos_individuales
    df_unificado['CREADOR'] = 'IMPORTADOR AUTOMATICO'
    df_unificado['FECHA CREACION'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    df_unificado['TRAZABILIDAD O MOVIMIENTOS'] = 'IMPORTACION AUTOMATICA'
    
    # Guardar archivo
    print(f"\\n✓ Guardando archivo: {ARCHIVO_SALIDA}")
    df_unificado.to_excel(ARCHIVO_SALIDA, index=False, sheet_name='INVENTARIO GENERAL')
    
    print("\\n" + "="*80)
    print("✓ IMPORTACIÓN COMPLETADA EXITOSAMENTE")
    print("="*80)
    print(f"\\nTotal de registros: {len(df_unificado)}")
    print(f"Archivo generado: {ARCHIVO_SALIDA}")
    
    # Resumen
    print(f"\\n📊 RESUMEN:")
    print(f"  • Archivos procesados: {len(dfs_procesados)}")
    print(f"  • Registros totales: {len(df_unificado)}")
    print(f"  • Sedes: {df_unificado['SEDE'].nunique()}")
    print(f"  • Tipos de tecnología: {df_unificado['TECNOLOGIA'].nunique()}")

if __name__ == "__main__":
    main()
'''

# Guardar el script
with open('/workspace/importador_inventarios.py', 'w', encoding='utf-8') as f:
    f.write(script_importador)

print("✓ Script de importación universal creado: importador_inventarios.py")
print("\nEste script puede ser ejecutado para importar automáticamente")
print("todos los archivos de la carpeta especificada.")
````
