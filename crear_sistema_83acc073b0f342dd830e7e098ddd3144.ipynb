````python

# CREAR IMPORTADOR UNIVERSAL
script_importador = '''"""
================================================================================
IMPORTADOR UNIVERSAL DE INVENTARIOS
================================================================================
Este script procesa mÃºltiples archivos Excel/CSV y los consolida en el
Sistema de Inventario Maestro.

CaracterÃ­sticas:
- Mapeo inteligente de columnas
- GeneraciÃ³n automÃ¡tica de cÃ³digos (SEDE, UNIFICADO, INDIVIDUAL)
- NormalizaciÃ³n de datos
- DetecciÃ³n automÃ¡tica de encabezados
- Procesamiento por lotes para archivos grandes
- Compatible con Power BI, SQL, PHP, Python

Autor: Sistema Automatizado
Fecha: 2025-01-20
================================================================================
"""

import pandas as pd
import numpy as np
import os
import glob
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# CONFIGURACIÃ“N
# ============================================================================

# Carpeta donde estÃ¡n los archivos a importar
CARPETA_ORIGEN = r"C:\\Users\\anderson.a\\Documents\\Inventario para importar"

# Archivo de salida
ARCHIVO_SALIDA = "SISTEMA_INVENTARIO_MAESTRO_CONSOLIDADO.xlsx"

# TamaÃ±o de lote para procesamiento (ajustar segÃºn memoria disponible)
CHUNK_SIZE = 1000

# ============================================================================
# MAPEO DE COLUMNAS
# ============================================================================

MAPEO_COLUMNAS = {
    'SERIAL': ['SERIAL', 'SERIE', 'SERIE No.', 'SERIE NO', 'SERIAL No.', 'NUMERO DE SERIE', 'No. SERIE'],
    'PLACA': ['PLACA', 'PLACA No.', 'PLACA NO', 'NUMERO PLACA', 'No. PLACA'],
    'SEDE': ['SEDE', 'CIUDAD', 'UBICACION', 'LOCATION', 'SITE'],
    'TECNOLOGIA': ['TECNOLOGIA', 'TIPO', 'DESCRIPCION', 'DESCRIPCION DE LOS ACTIVOS', 'TIPO DE EQUIPO', 'TIPO EQUIPO'],
    'MARCA': ['MARCA', 'BRAND', 'FABRICANTE'],
    'MODELO': ['MODELO', 'MODEL'],
    'PROCESADOR': ['PROCESADOR', 'PROCESSOR', 'CPU'],
    'RAM': ['RAM', 'CANTIDAD RAM', 'MEMORIA', 'MEMORIA RAM', 'CANTIDAD DE RAM'],
    'ARQUITECTURA_RAM': ['ARQUITECTURA RAM', 'ARQUI RAM', 'TIPO RAM', 'TIPO DE RAM'],
    'TIPO_DISCO': ['TIPO DE DISCO', 'TIPO DISCO', 'DISK TYPE'],
    'ALMACENAMIENTO': ['ALMACENAMIENTO', 'ESPACIO DISCO', 'DISCO', 'STORAGE', 'CAPACIDAD', 'ESPACIO DE DISCO'],
    'SO': ['SO', 'SISTEMA OPERATIVO', 'OS', 'OPERATING SYSTEM'],
    'MAC': ['MAC', 'MAC ADDRESS', 'DIRECCION MAC'],
    'IP': ['IP', 'IP ADDRESS', 'DIRECCION IP'],
    'HOSTNAME': ['HOSTNAME', 'HOST NAME', 'NOMBRE EQUIPO'],
    'ESTADO': ['ESTADO', 'ESTADO ACTUAL', 'STATUS', 'CONDICION'],
    'ASIGNADO_A': ['ASIGNADO A', 'ASIGNADO NUEVO', 'USUARIO', 'RESPONSABLE', 'ASIGNADO', 'PERSONA ASIGNADA'],
    'ANTERIOR_ASIGNADO': ['ANTERIOR ASIGNADO', 'USUARIO ANTERIOR', 'ASIGNADO ANTERIOR'],
    'CARGO': ['CARGO', 'PUESTO', 'POSITION'],
    'AREA': ['AREA', 'DEPARTAMENTO', 'DEPT', 'DEPARTMENT'],
    'CONTACTO': ['CONTACTO', 'TELEFONO', 'EMAIL', 'CORREO'],
    'FECHA_ASIGNACION': ['FECHA ASIGNACION', 'FECHA', 'FECHA DE ASIGNACION'],
    'FECHA_LLEGADA': ['FECHA DE LLEGADA', 'FECHA LLEGADA', 'FECHA INGRESO'],
    'OBSERVACIONES': ['OBSERVACIONES', 'OBSERVACION', 'NOTAS', 'COMENTARIOS'],
    'DISPONIBLE': ['DISPONIBLE', 'DISPONIBILIDAD', 'AVAILABLE'],
    'ENTRADA_OC': ['ENTRADA OC COMPRA', 'OC COMPRA', 'ORDEN COMPRA', 'ENTRADA OC'],
    'ENVIADO_CARGADO': ['ENVIADO /CARGADO', 'ENVIADO', 'CARGADO'],
    'OC': ['OC', 'ORDEN', 'ORDEN DE COMPRA'],
    'PROVEEDOR': ['PROVEEDOR', 'SUPPLIER', 'VENDOR'],
    'MONITOR_SERIAL': ['SERIAL MONITOR', 'SERIE MONITOR', 'SERIAL2', 'MONITOR SERIAL'],
    'MONITOR_PLACA': ['PLACA MONITOR', 'PLACA DE MONITOR'],
    'MONITOR_MARCA': ['MARCA DE MONITOR', 'MARCA MONITOR'],
    'MONITOR_MODELO': ['MODELO MONITOR', 'MODELO DE MONITOR'],
    'MOUSE': ['MOUSE', 'RATON'],
    'TECLADO': ['TECLADO', 'KEYBOARD']
}

# ============================================================================
# FUNCIONES AUXILIARES
# ============================================================================

def normalizar_texto(texto):
    """Normaliza texto: elimina espacios extras, convierte a mayÃºsculas"""
    if pd.isna(texto):
        return ''
    return str(texto).strip().upper()

def mapear_columna(df, posibles_nombres):
    """Busca una columna en el DataFrame usando mÃºltiples nombres posibles"""
    for nombre in posibles_nombres:
        # Buscar coincidencia exacta (case insensitive)
        for col in df.columns:
            if str(col).strip().upper() == nombre.upper():
                return col
        # Buscar coincidencia parcial
        for col in df.columns:
            if nombre.upper() in str(col).strip().upper():
                return col
    return None

def detectar_fila_encabezados(df, max_rows=10):
    """Detecta automÃ¡ticamente la fila de encabezados"""
    for i in range(min(max_rows, len(df))):
        row = df.iloc[i]
        # Contar cuÃ¡ntas celdas no vacÃ­as hay
        non_empty = row.notna().sum()
        # Si mÃ¡s del 50% de las celdas tienen datos, probablemente es el encabezado
        if non_empty > len(row) * 0.5:
            # Verificar si contiene palabras clave de encabezados
            row_str = ' '.join(row.astype(str).str.upper())
            keywords = ['SERIAL', 'PLACA', 'MARCA', 'MODELO', 'SEDE', 'TECNOLOGIA']
            if any(kw in row_str for kw in keywords):
                return i
    return 0

def generar_codigo_sede(sede):
    """Genera cÃ³digo de 3 letras para la sede"""
    if pd.isna(sede) or sede == '':
        return 'GEN'
    
    sede_upper = str(sede).strip().upper()
    mapeo_sedes = {
        'MEDELLIN': 'MED', 'CARTAGENA': 'CTG', 'PASTO': 'PAS',
        'BOGOTA': 'BOG', 'CALI': 'CAL', 'BARRANQUILLA': 'BAQ',
        'BUCARAMANGA': 'BUC', 'PEREIRA': 'PER', 'MANIZALES': 'MAN',
        'CUCUTA': 'CUC', 'IBAGUE': 'IBA', 'TUNJA': 'TUN',
        'MONTERIA': 'MON', 'VILLAVICENCIO': 'VIL'
    }
    
    for ciudad, codigo in mapeo_sedes.items():
        if ciudad in sede_upper:
            return codigo
    
    return sede_upper[:3]

def generar_codigo_tecnologia(tecnologia):
    """Genera cÃ³digo abreviado para tipo de tecnologÃ­a"""
    if pd.isna(tecnologia) or tecnologia == '':
        return 'GEN'
    
    tech_upper = str(tecnologia).strip().upper()
    mapeo_tech = {
        'PORTATIL': 'PORT', 'LAPTOP': 'PORT', 'NOTEBOOK': 'PORT',
        'PC ESCRITORIO': 'DESK', 'DESKTOP': 'DESK', 'EQUIPO DE COMPUTO': 'COMP',
        'TODO EN UNO': 'AIO', 'ALL IN ONE': 'AIO',
        'MONITOR': 'MON', 'PANTALLA': 'MON',
        'TECLADO': 'TEC', 'KEYBOARD': 'TEC',
        'MOUSE': 'MOU', 'RATON': 'MOU',
        'IMPRESORA': 'IMP', 'PRINTER': 'IMP',
        'TELEFONO': 'TEL', 'PHONE': 'TEL',
        'SWITCH': 'SWI', 'ROUTER': 'ROU',
        'SERVIDOR': 'SRV', 'SERVER': 'SRV',
        'TABLET': 'TAB', 'CELULAR': 'CEL', 'MOVIL': 'CEL',
        'CCTV': 'CAM', 'CAMARA': 'CAM',
        'ESCANER': 'ESC', 'SCANNER': 'ESC',
        'UPS': 'UPS', 'BASE': 'BAS', 'DIADEMA': 'DIA'
    }
    
    for tech, codigo in mapeo_tech.items():
        if tech in tech_upper:
            return codigo
    
    return tech_upper[:4]

def cargar_archivo(filepath):
    """Carga un archivo Excel o CSV detectando automÃ¡ticamente los encabezados"""
    print(f"\\n  ðŸ“„ Procesando: {os.path.basename(filepath)}")
    
    try:
        # Intentar leer como Excel
        if filepath.endswith(('.xlsx', '.xls')):
            # Leer sin encabezados primero
            df_temp = pd.read_excel(filepath, header=None, nrows=20)
            header_row = detectar_fila_encabezados(df_temp)
            
            # Cargar con el encabezado correcto
            df = pd.read_excel(filepath, header=header_row)
            
            # Si los encabezados son "Unnamed", usar la primera fila como encabezados
            if 'Unnamed' in str(df.columns[0]):
                df.columns = df.iloc[0]
                df = df[1:].reset_index(drop=True)
        else:
            # CSV
            df = pd.read_csv(filepath, encoding='utf-8')
        
        print(f"     âœ“ {len(df)} registros cargados")
        return df
        
    except Exception as e:
        print(f"     âœ— Error: {str(e)}")
        return pd.DataFrame()

def procesar_archivo(df, nombre_archivo):
    """Procesa un DataFrame y lo mapea a la estructura del inventario"""
    
    # Crear DataFrame con estructura estÃ¡ndar
    df_procesado = pd.DataFrame()
    
    # Mapear cada columna
    columnas_mapeadas = 0
    for col_destino, posibles_nombres in MAPEO_COLUMNAS.items():
        col_fuente = mapear_columna(df, posibles_nombres)
        if col_fuente:
            df_procesado[col_destino] = df[col_fuente].apply(normalizar_texto)
            columnas_mapeadas += 1
    
    print(f"     âœ“ {columnas_mapeadas} columnas mapeadas")
    
    return df_procesado

def generar_codigos(df):
    """Genera cÃ³digos automÃ¡ticos para el inventario"""
    print("\\n  ðŸ”¢ Generando cÃ³digos automÃ¡ticos...")
    
    # ID secuencial
    df['ID'] = range(1, len(df) + 1)
    
    # CÃ³digos individuales y unificados
    codigos_individuales = []
    codigos_unificados = []
    codigos_sede_global = []
    
    contadores = {}
    contadores_sede = {}
    
    for idx, row in df.iterrows():
        sede = row.get('SEDE', '')
        tecnologia = row.get('TECNOLOGIA', '')
        placa_existente = row.get('PLACA', '')
        
        codigo_sede = generar_codigo_sede(sede)
        codigo_tech = generar_codigo_tecnologia(tecnologia)
        
        # CÃ³digo individual
        if placa_existente and placa_existente.strip() != '':
            cod_individual = placa_existente
        else:
            clave = f"{codigo_sede}_{codigo_tech}"
            contadores[clave] = contadores.get(clave, 0) + 1
            numero = str(contadores[clave]).zfill(3)
            cod_individual = f"{codigo_sede}-{codigo_tech}-{numero}"
        
        codigos_individuales.append(cod_individual)
        
        # CÃ³digo unificado (por ahora igual al individual)
        # TODO: Implementar lÃ³gica de agrupaciÃ³n
        codigos_unificados.append(cod_individual)
        
        # CÃ³digo de sede global
        contadores_sede[codigo_sede] = contadores_sede.get(codigo_sede, 0) + 1
        numero_sede = str(contadores_sede[codigo_sede]).zfill(4)
        cod_sede = f"{codigo_sede}-{numero_sede}"
        codigos_sede_global.append(cod_sede)
    
    df['CODIGO_INDIVIDUAL'] = codigos_individuales
    df['CODIGO_UNIFICADO'] = codigos_unificados
    df['CODIGO_SEDE'] = codigos_sede_global
    
    print(f"     âœ“ {len(df)} cÃ³digos generados")
    
    return df

# ============================================================================
# PROCESO PRINCIPAL
# ============================================================================

def main():
    print("="*80)
    print("IMPORTADOR UNIVERSAL DE INVENTARIOS")
    print("="*80)
    print(f"\\nFecha: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    
    # Verificar carpeta
    if not os.path.exists(CARPETA_ORIGEN):
        print(f"\\nâœ— ERROR: No se encuentra la carpeta: {CARPETA_ORIGEN}")
        return
    
    # Buscar archivos
    archivos = []
    for ext in ['*.xlsx', '*.xls', '*.csv']:
        archivos.extend(glob.glob(os.path.join(CARPETA_ORIGEN, ext)))
    
    print(f"\\nâœ“ Encontrados {len(archivos)} archivos para procesar")
    
    # Procesar archivos
    dfs_procesados = []
    
    for archivo in archivos:
        df = cargar_archivo(archivo)
        if len(df) > 0:
            df_proc = procesar_archivo(df, os.path.basename(archivo))
            if len(df_proc) > 0:
                dfs_procesados.append(df_proc)
    
    if len(dfs_procesados) == 0:
        print("\\nâœ— No se procesaron archivos exitosamente")
        return
    
    # Consolidar
    print(f"\\nðŸ“Š CONSOLIDANDO DATOS...")
    df_consolidado = pd.concat(dfs_procesados, ignore_index=True)
    print(f"   âœ“ {len(df_consolidado)} registros consolidados")
    
    # Generar cÃ³digos
    df_consolidado = generar_codigos(df_consolidado)
    
    # Agregar metadatos
    df_consolidado['CREADOR'] = 'IMPORTADOR AUTOMATICO'
    df_consolidado['FECHA_CREACION'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    df_consolidado['TRAZABILIDAD'] = 'ImportaciÃ³n automÃ¡tica'
    
    # Guardar
    print(f"\\nðŸ’¾ GUARDANDO ARCHIVO...")
    df_consolidado.to_excel(ARCHIVO_SALIDA, index=False, sheet_name='INVENTARIO_MAESTRO')
    
    print("\\n" + "="*80)
    print("âœ“ IMPORTACIÃ“N COMPLETADA EXITOSAMENTE")
    print("="*80)
    print(f"\\nArchivo generado: {ARCHIVO_SALIDA}")
    print(f"Total de registros: {len(df_consolidado)}")
    print(f"Archivos procesados: {len(dfs_procesados)}")
    
    # Resumen
    print(f"\\nðŸ“Š RESUMEN:")
    if 'SEDE' in df_consolidado.columns:
        print(f"\\n  Sedes:")
        for sede, count in df_consolidado['SEDE'].value_counts().head(10).items():
            print(f"    â€¢ {sede}: {count} equipos")
    
    if 'TECNOLOGIA' in df_consolidado.columns:
        print(f"\\n  TecnologÃ­as:")
        for tech, count in df_consolidado['TECNOLOGIA'].value_counts().head(10).items():
            print(f"    â€¢ {tech}: {count} unidades")
    
    if 'ESTADO' in df_consolidado.columns:
        print(f"\\n  Estados:")
        for estado, count in df_consolidado['ESTADO'].value_counts().items():
            print(f"    â€¢ {estado}: {count} equipos")

if __name__ == "__main__":
    main()
'''

# Guardar el script
with open('/workspace/importador_universal.py', 'w', encoding='utf-8') as f:
    f.write(script_importador)

print("\nâœ“ IMPORTADOR UNIVERSAL creado: importador_universal.py")
print("  Este script procesarÃ¡ todos los archivos de la carpeta especificada")
````

````python

# CREAR API REST EN PHP
script_php = '''<?php
/**
 * ============================================================================
 * API REST PARA INVENTARIO
 * ============================================================================
 * Proporciona endpoints para consultar y gestionar el inventario
 * 
 * Endpoints:
 * - GET /api/inventario - Listar todos los equipos
 * - GET /api/inventario/{id} - Obtener equipo por ID
 * - GET /api/inventario/serial/{serial} - Buscar por serial
 * - GET /api/inventario/sede/{sede} - Filtrar por sede
 * - POST /api/inventario - Crear nuevo equipo
 * - PUT /api/inventario/{id} - Actualizar equipo
 * - DELETE /api/inventario/{id} - Eliminar equipo
 * ============================================================================
 */

header('Content-Type: application/json; charset=utf-8');
header('Access-Control-Allow-Origin: *');
header('Access-Control-Allow-Methods: GET, POST, PUT, DELETE');
header('Access-Control-Allow-Headers: Content-Type');

// ConfiguraciÃ³n de base de datos
$host = 'localhost';
$dbname = 'inventario_maestro';
$username = 'root';
$password = '';

try {
    $pdo = new PDO("mysql:host=$host;dbname=$dbname;charset=utf8", $username, $password);
    $pdo->setAttribute(PDO::ATTR_ERRMODE, PDO::ERRMODE_EXCEPTION);
} catch(PDOException $e) {
    http_response_code(500);
    echo json_encode(['error' => 'Error de conexiÃ³n: ' . $e->getMessage()]);
    exit;
}

// Obtener mÃ©todo y ruta
$method = $_SERVER['REQUEST_METHOD'];
$request = $_SERVER['REQUEST_URI'];
$path = parse_url($request, PHP_URL_PATH);
$segments = explode('/', trim($path, '/'));

// Routing
if ($segments[0] === 'api' && $segments[1] === 'inventario') {
    
    switch($method) {
        case 'GET':
            if (isset($segments[2])) {
                if (is_numeric($segments[2])) {
                    // GET por ID
                    obtenerPorId($pdo, $segments[2]);
                } elseif ($segments[2] === 'serial' && isset($segments[3])) {
                    // GET por serial
                    buscarPorSerial($pdo, $segments[3]);
                } elseif ($segments[2] === 'sede' && isset($segments[3])) {
                    // GET por sede
                    filtrarPorSede($pdo, $segments[3]);
                }
            } else {
                // GET todos
                listarTodos($pdo);
            }
            break;
            
        case 'POST':
            crearEquipo($pdo);
            break;
            
        case 'PUT':
            if (isset($segments[2])) {
                actualizarEquipo($pdo, $segments[2]);
            }
            break;
            
        case 'DELETE':
            if (isset($segments[2])) {
                eliminarEquipo($pdo, $segments[2]);
            }
            break;
    }
}

// ============================================================================
// FUNCIONES
// ============================================================================

function listarTodos($pdo) {
    $stmt = $pdo->query("SELECT * FROM inventario ORDER BY id DESC");
    $equipos = $stmt->fetchAll(PDO::FETCH_ASSOC);
    echo json_encode([
        'success' => true,
        'total' => count($equipos),
        'data' => $equipos
    ]);
}

function obtenerPorId($pdo, $id) {
    $stmt = $pdo->prepare("SELECT * FROM inventario WHERE id = ?");
    $stmt->execute([$id]);
    $equipo = $stmt->fetch(PDO::FETCH_ASSOC);
    
    if ($equipo) {
        echo json_encode(['success' => true, 'data' => $equipo]);
    } else {
        http_response_code(404);
        echo json_encode(['success' => false, 'error' => 'Equipo no encontrado']);
    }
}

function buscarPorSerial($pdo, $serial) {
    $stmt = $pdo->prepare("SELECT * FROM inventario WHERE serial LIKE ?");
    $stmt->execute(["%$serial%"]);
    $equipos = $stmt->fetchAll(PDO::FETCH_ASSOC);
    echo json_encode(['success' => true, 'total' => count($equipos), 'data' => $equipos]);
}

function filtrarPorSede($pdo, $sede) {
    $stmt = $pdo->prepare("SELECT * FROM inventario WHERE sede = ?");
    $stmt->execute([$sede]);
    $equipos = $stmt->fetchAll(PDO::FETCH_ASSOC);
    echo json_encode(['success' => true, 'total' => count($equipos), 'data' => $equipos]);
}

function crearEquipo($pdo) {
    $data = json_decode(file_get_contents('php://input'), true);
    
    $sql = "INSERT INTO inventario (codigo_individual, serial, placa, sede, tecnologia, marca, modelo, estado, creador, fecha_creacion) 
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, NOW())";
    
    try {
        $stmt = $pdo->prepare($sql);
        $stmt->execute([
            $data['codigo_individual'],
            $data['serial'],
            $data['placa'],
            $data['sede'],
            $data['tecnologia'],
            $data['marca'],
            $data['modelo'],
            $data['estado'],
            $data['creador']
        ]);
        
        echo json_encode(['success' => true, 'id' => $pdo->lastInsertId()]);
    } catch(PDOException $e) {
        http_response_code(400);
        echo json_encode(['success' => false, 'error' => $e->getMessage()]);
    }
}

function actualizarEquipo($pdo, $id) {
    $data = json_decode(file_get_contents('php://input'), true);
    
    $sql = "UPDATE inventario SET 
            asignado_a = ?, 
            cargo = ?, 
            area = ?, 
            estado = ?,
            observaciones = ?
            WHERE id = ?";
    
    try {
        $stmt = $pdo->prepare($sql);
        $stmt->execute([
            $data['asignado_a'],
            $data['cargo'],
            $data['area'],
            $data['estado'],
            $data['observaciones'],
            $id
        ]);
        
        echo json_encode(['success' => true, 'message' => 'Equipo actualizado']);
    } catch(PDOException $e) {
        http_response_code(400);
        echo json_encode(['success' => false, 'error' => $e->getMessage()]);
    }
}

function eliminarEquipo($pdo, $id) {
    try {
        $stmt = $pdo->prepare("DELETE FROM inventario WHERE id = ?");
        $stmt->execute([$id]);
        echo json_encode(['success' => true, 'message' => 'Equipo eliminado']);
    } catch(PDOException $e) {
        http_response_code(400);
        echo json_encode(['success' => false, 'error' => $e->getMessage()]);
    }
}
?>
'''

with open('/workspace/api_inventario.php', 'w', encoding='utf-8') as f:
    f.write(script_php)

print("âœ“ API REST en PHP creada: api_inventario.php")
````

````python

# CREAR MÃ“DULO PYTHON PARA GESTIÃ“N
script_python = '''"""
================================================================================
MÃ“DULO PYTHON PARA GESTIÃ“N DE INVENTARIO
================================================================================
Proporciona clases y funciones para gestionar el inventario programÃ¡ticamente
"""

import pandas as pd
import sqlite3
from datetime import datetime
from typing import List, Dict, Optional

class GestorInventario:
    """Clase principal para gestionar el inventario"""
    
    def __init__(self, archivo_excel: str = None, db_path: str = 'inventario.db'):
        """
        Inicializa el gestor de inventario
        
        Args:
            archivo_excel: Ruta al archivo Excel del inventario
            db_path: Ruta a la base de datos SQLite
        """
        self.archivo_excel = archivo_excel
        self.db_path = db_path
        self.df = None
        
        if archivo_excel:
            self.cargar_desde_excel()
        else:
            self.cargar_desde_db()
    
    def cargar_desde_excel(self):
        """Carga el inventario desde archivo Excel"""
        try:
            self.df = pd.read_excel(self.archivo_excel, sheet_name='INVENTARIO MAESTRO')
            print(f"âœ“ Inventario cargado: {len(self.df)} registros")
        except Exception as e:
            print(f"âœ— Error al cargar Excel: {e}")
            self.df = pd.DataFrame()
    
    def cargar_desde_db(self):
        """Carga el inventario desde base de datos SQLite"""
        try:
            conn = sqlite3.connect(self.db_path)
            self.df = pd.read_sql_query("SELECT * FROM inventario", conn)
            conn.close()
            print(f"âœ“ Inventario cargado desde DB: {len(self.df)} registros")
        except Exception as e:
            print(f"âœ— Error al cargar DB: {e}")
            self.df = pd.DataFrame()
    
    def buscar_por_serial(self, serial: str) -> pd.DataFrame:
        """Busca equipos por serial"""
        if self.df is None or len(self.df) == 0:
            return pd.DataFrame()
        
        resultado = self.df[self.df['SERIAL'].str.contains(serial, case=False, na=False)]
        return resultado
    
    def buscar_por_placa(self, placa: str) -> pd.DataFrame:
        """Busca equipos por placa"""
        if self.df is None or len(self.df) == 0:
            return pd.DataFrame()
        
        resultado = self.df[self.df['PLACA'].str.contains(placa, case=False, na=False)]
        return resultado
    
    def filtrar_por_sede(self, sede: str) -> pd.DataFrame:
        """Filtra equipos por sede"""
        if self.df is None or len(self.df) == 0:
            return pd.DataFrame()
        
        resultado = self.df[self.df['SEDE'].str.upper() == sede.upper()]
        return resultado
    
    def filtrar_por_estado(self, estado: str) -> pd.DataFrame:
        """Filtra equipos por estado"""
        if self.df is None or len(self.df) == 0:
            return pd.DataFrame()
        
        resultado = self.df[self.df['ESTADO'].str.upper() == estado.upper()]
        return resultado
    
    def filtrar_por_usuario(self, usuario: str) -> pd.DataFrame:
        """Filtra equipos asignados a un usuario"""
        if self.df is None or len(self.df) == 0:
            return pd.DataFrame()
        
        resultado = self.df[self.df['ASIGNADO_A'].str.contains(usuario, case=False, na=False)]
        return resultado
    
    def asignar_equipo(self, codigo_individual: str, usuario: str, cargo: str, 
                      area: str, observaciones: str = '') -> bool:
        """
        Asigna un equipo a un usuario
        
        Args:
            codigo_individual: CÃ³digo del equipo
            usuario: Nombre del usuario
            cargo: Cargo del usuario
            area: Ãrea del usuario
            observaciones: Observaciones adicionales
        
        Returns:
            True si la asignaciÃ³n fue exitosa
        """
        try:
            idx = self.df[self.df['CODIGO_INDIVIDUAL'] == codigo_individual].index
            
            if len(idx) == 0:
                print(f"âœ— Equipo no encontrado: {codigo_individual}")
                return False
            
            # Guardar usuario anterior
            usuario_anterior = self.df.loc[idx[0], 'ASIGNADO_A']
            self.df.loc[idx[0], 'ANTERIOR_ASIGNADO'] = usuario_anterior
            
            # Asignar nuevo usuario
            self.df.loc[idx[0], 'ASIGNADO_A'] = usuario
            self.df.loc[idx[0], 'CARGO'] = cargo
            self.df.loc[idx[0], 'AREA'] = area
            self.df.loc[idx[0], 'FECHA_ASIGNACION'] = datetime.now().strftime('%Y-%m-%d')
            self.df.loc[idx[0], 'ESTADO'] = 'ACTIVO'
            self.df.loc[idx[0], 'DISPONIBLE'] = 'NO'
            
            # Actualizar trazabilidad
            trazabilidad_actual = self.df.loc[idx[0], 'TRAZABILIDAD']
            nueva_trazabilidad = f"{trazabilidad_actual} | Asignado a {usuario} el {datetime.now().strftime('%Y-%m-%d')}"
            self.df.loc[idx[0], 'TRAZABILIDAD'] = nueva_trazabilidad
            
            if observaciones:
                self.df.loc[idx[0], 'OBSERVACIONES'] = observaciones
            
            print(f"âœ“ Equipo {codigo_individual} asignado a {usuario}")
            return True
            
        except Exception as e:
            print(f"âœ— Error al asignar equipo: {e}")
            return False
    
    def dar_de_baja(self, codigo_individual: str, motivo: str) -> bool:
        """Da de baja un equipo"""
        try:
            idx = self.df[self.df['CODIGO_INDIVIDUAL'] == codigo_individual].index
            
            if len(idx) == 0:
                print(f"âœ— Equipo no encontrado: {codigo_individual}")
                return False
            
            self.df.loc[idx[0], 'ESTADO'] = 'DADO DE BAJA'
            self.df.loc[idx[0], 'DISPONIBLE'] = 'NO'
            self.df.loc[idx[0], 'OBSERVACIONES'] = f"BAJA: {motivo}"
            
            # Actualizar trazabilidad
            trazabilidad_actual = self.df.loc[idx[0], 'TRAZABILIDAD']
            nueva_trazabilidad = f"{trazabilidad_actual} | Dado de baja el {datetime.now().strftime('%Y-%m-%d')}: {motivo}"
            self.df.loc[idx[0], 'TRAZABILIDAD'] = nueva_trazabilidad
            
            print(f"âœ“ Equipo {codigo_individual} dado de baja")
            return True
            
        except Exception as e:
            print(f"âœ— Error al dar de baja: {e}")
            return False
    
    def generar_reporte_sede(self, sede: str) -> Dict:
        """Genera reporte estadÃ­stico de una sede"""
        df_sede = self.filtrar_por_sede(sede)
        
        if len(df_sede) == 0:
            return {}
        
        reporte = {
            'sede': sede,
            'total_equipos': len(df_sede),
            'activos': len(df_sede[df_sede['ESTADO'] == 'ACTIVO']),
            'disponibles': len(df_sede[df_sede['DISPONIBLE'] == 'SI']),
            'dados_baja': len(df_sede[df_sede['ESTADO'] == 'DADO DE BAJA']),
            'por_tecnologia': df_sede['TECNOLOGIA'].value_counts().to_dict(),
            'por_marca': df_sede['MARCA'].value_counts().to_dict()
        }
        
        return reporte
    
    def guardar_excel(self, archivo_salida: str = None):
        """Guarda el inventario en Excel"""
        if archivo_salida is None:
            archivo_salida = self.archivo_excel
        
        try:
            self.df.to_excel(archivo_salida, sheet_name='INVENTARIO MAESTRO', index=False)
            print(f"âœ“ Inventario guardado en: {archivo_salida}")
        except Exception as e:
            print(f"âœ— Error al guardar: {e}")
    
    def guardar_db(self):
        """Guarda el inventario en base de datos SQLite"""
        try:
            conn = sqlite3.connect(self.db_path)
            self.df.to_sql('inventario', conn, if_exists='replace', index=False)
            conn.close()
            print(f"âœ“ Inventario guardado en DB: {self.db_path}")
        except Exception as e:
            print(f"âœ— Error al guardar en DB: {e}")


# ============================================================================
# EJEMPLO DE USO
# ============================================================================

if __name__ == "__main__":
    # Crear instancia del gestor
    gestor = GestorInventario("SISTEMA_INVENTARIO_MAESTRO_CONSOLIDADO.xlsx")
    
    # Buscar por serial
    print("\\n=== BUSCAR POR SERIAL ===")
    resultado = gestor.buscar_por_serial("MP2B7YWL")
    if len(resultado) > 0:
        print(resultado[['CODIGO_INDIVIDUAL', 'SERIAL', 'MARCA', 'MODELO', 'ASIGNADO_A']])
    
    # Filtrar por sede
    print("\\n=== EQUIPOS EN MEDELLÃN ===")
    medellin = gestor.filtrar_por_sede("MEDELLIN")
    print(f"Total: {len(medellin)} equipos")
    
    # Generar reporte
    print("\\n=== REPORTE DE SEDE ===")
    reporte = gestor.generar_reporte_sede("MEDELLIN")
    for key, value in reporte.items():
        print(f"{key}: {value}")
    
    # Asignar equipo
    print("\\n=== ASIGNAR EQUIPO ===")
    gestor.asignar_equipo(
        codigo_individual="MED-PORT-001",
        usuario="JUAN PEREZ",
        cargo="ANALISTA",
        area="SISTEMAS",
        observaciones="AsignaciÃ³n inicial"
    )
    
    # Guardar cambios
    gestor.guardar_excel("INVENTARIO_ACTUALIZADO.xlsx")
'''

with open('/workspace/gestor_inventario.py', 'w', encoding='utf-8') as f:
    f.write(script_python)

print("âœ“ MÃ³dulo Python creado: gestor_inventario.py")
````
