````python

# CREAR IMPORTADOR MASIVO ADAPTADO A TU BD
importador_workmanager = '''"""
IMPORTADOR MASIVO PARA WORKMANAGER ERP
Llena la tabla equipos_individuales sin l√≠mites
Compatible con tu estructura existente
"""

import pandas as pd
import sqlite3
from datetime import datetime
import os
import glob

# ============================================================================
# CONFIGURACI√ìN
# ============================================================================

DB_PATH = 'workmanager_erp.db'
CARPETA_ORIGEN = r"C:\\Users\\anderson.a\\Documents\\Inventario para importar"
CHUNK_SIZE = 500

# ============================================================================
# MAPEO A TU TABLA equipos_individuales
# ============================================================================

MAPEO_COLUMNAS = {
    'codigo_barras': ['CODIGO', 'CODIGO BARRAS', 'CODIGO_BARRAS', 'CODIGO INDIVIDUAL', 'PLACA'],
    'tecnologia': ['TECNOLOGIA', 'TIPO', 'TIPO DE EQUIPO', 'DESCRIPCION'],
    'serial': ['SERIAL', 'SERIE', 'NUMERO DE SERIE', 'No. SERIE'],
    'placa': ['PLACA', 'PLACA No.', 'NUMERO PLACA'],
    'marca': ['MARCA', 'FABRICANTE'],
    'modelo': ['MODELO', 'MODEL'],
    'procesador': ['PROCESADOR', 'CPU', 'PROCESSOR'],
    'ram': ['RAM', 'MEMORIA', 'CANTIDAD RAM'],
    'disco': ['DISCO', 'ALMACENAMIENTO', 'STORAGE', 'ESPACIO DISCO'],
    'so': ['SO', 'SISTEMA OPERATIVO', 'OS'],
    'ip': ['IP', 'IP ADDRESS', 'DIRECCION IP'],
    'mac': ['MAC', 'MAC ADDRESS'],
    'hostname': ['HOSTNAME', 'NOMBRE EQUIPO'],
    'asignado_a': ['ASIGNADO A', 'USUARIO', 'ASIGNADO'],
    'sede_id': ['SEDE', 'CIUDAD', 'UBICACION'],
    'estado': ['ESTADO', 'STATUS'],
    'disponible': ['DISPONIBLE', 'DISPONIBILIDAD'],
    'observaciones': ['OBSERVACIONES', 'NOTAS', 'COMENTARIOS'],
    'mouse': ['MOUSE', 'RATON'],
    'teclado': ['TECLADO', 'KEYBOARD'],
    'monitor_marca': ['MARCA MONITOR', 'MARCA DE MONITOR'],
    'monitor_modelo': ['MODELO MONITOR', 'MODELO DE MONITOR'],
    'monitor_serial': ['SERIAL MONITOR', 'SERIE MONITOR'],
    'telefono_marca': ['MARCA TELEFONO', 'TELEFONO MARCA'],
    'telefono_serial': ['SERIAL TELEFONO', 'TELEFONO SERIAL'],
    'impresora_marca': ['MARCA IMPRESORA', 'IMPRESORA MARCA'],
    'impresora_ip': ['IP IMPRESORA', 'IMPRESORA IP'],
    'cctv_marca': ['MARCA CCTV', 'CCTV MARCA']
}

# ============================================================================
# FUNCIONES
# ============================================================================

def normalizar_texto(texto):
    if pd.isna(texto):
        return None
    return str(texto).strip().upper() if str(texto).strip() else None

def mapear_columna(df, posibles_nombres):
    for nombre in posibles_nombres:
        for col in df.columns:
            if str(col).strip().upper() == nombre.upper():
                return col
            if nombre.upper() in str(col).strip().upper():
                return col
    return None

def detectar_encabezados(df, max_rows=15):
    for i in range(min(max_rows, len(df))):
        row = df.iloc[i]
        non_empty = row.notna().sum()
        if non_empty > len(row) * 0.4:
            row_str = ' '.join(row.astype(str).str.upper())
            keywords = ['SERIAL', 'PLACA', 'MARCA', 'MODELO', 'TECNOLOGIA']
            if any(kw in row_str for kw in keywords):
                return i
    return 0

def obtener_sede_id(nombre_sede, conn):
    """Obtiene ID de sede desde tu tabla sedes"""
    if not nombre_sede:
        return None
    
    cursor = conn.cursor()
    cursor.execute("SELECT id FROM sedes WHERE nombre LIKE ? OR ciudad LIKE ?", 
                  (f'%{nombre_sede}%', f'%{nombre_sede}%'))
    result = cursor.fetchone()
    return result[0] if result else None

def generar_codigo_barras(tecnologia, contador):
    """Genera c√≥digo de barras √∫nico"""
    mapeo_tech = {
        'PORTATIL': 'PORT', 'LAPTOP': 'PORT',
        'PC ESCRITORIO': 'DESK', 'DESKTOP': 'DESK',
        'TODO EN UNO': 'AIO',
        'MONITOR': 'MON',
        'IMPRESORA': 'IMP',
        'TELEFONO': 'TEL'
    }
    
    tech_upper = str(tecnologia).upper() if tecnologia else 'GEN'
    codigo_tech = 'GEN'
    
    for tech, codigo in mapeo_tech.items():
        if tech in tech_upper:
            codigo_tech = codigo
            break
    
    return f"{codigo_tech}-{str(contador).zfill(4)}"

def cargar_archivo(filepath):
    print(f"\\n  üìÑ {os.path.basename(filepath)}")
    
    try:
        if filepath.endswith(('.xlsx', '.xls')):
            df_temp = pd.read_excel(filepath, header=None, nrows=20)
            header_row = detectar_encabezados(df_temp)
            df = pd.read_excel(filepath, header=header_row)
            
            if 'Unnamed' in str(df.columns[0]):
                df.columns = df.iloc[0]
                df = df[1:].reset_index(drop=True)
        else:
            df = pd.read_csv(filepath, encoding='utf-8')
        
        print(f"     ‚úì {len(df):,} registros")
        return df
    except Exception as e:
        print(f"     ‚úó Error: {e}")
        return pd.DataFrame()

def procesar_dataframe(df, conn):
    """Mapea DataFrame a equipos_individuales"""
    df_procesado = pd.DataFrame()
    
    for col_bd, posibles_nombres in MAPEO_COLUMNAS.items():
        col_fuente = mapear_columna(df, posibles_nombres)
        if col_fuente:
            df_procesado[col_bd] = df[col_fuente].apply(normalizar_texto)
    
    # Obtener sede_id
    if 'sede_id' in df_procesado.columns:
        df_procesado['sede_id'] = df_procesado['sede_id'].apply(
            lambda x: obtener_sede_id(x, conn) if x else None
        )
    
    return df_procesado

def main():
    print("="*80)
    print("IMPORTADOR MASIVO PARA WORKMANAGER ERP")
    print("="*80)
    print(f"\\nFecha: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    
    # Conectar
    conn = sqlite3.connect(DB_PATH)
    print(f"‚úì Conectado a {DB_PATH}")
    
    # Obtener √∫ltimo contador
    cursor = conn.cursor()
    cursor.execute("SELECT COUNT(*) FROM equipos_individuales")
    contador_inicial = cursor.fetchone()[0]
    
    # Buscar archivos
    if not os.path.exists(CARPETA_ORIGEN):
        print(f"\\n‚úó ERROR: Carpeta no encontrada: {CARPETA_ORIGEN}")
        return
    
    archivos = []
    for ext in ['*.xlsx', '*.xls', '*.csv']:
        archivos.extend(glob.glob(os.path.join(CARPETA_ORIGEN, ext)))
    
    print(f"\\nüìÇ Encontrados {len(archivos)} archivos")
    
    total_insertados = 0
    contador = contador_inicial + 1
    
    for archivo in archivos:
        try:
            df = cargar_archivo(archivo)
            if len(df) == 0:
                continue
            
            df_procesado = procesar_dataframe(df, conn)
            
            # Generar c√≥digos de barras si no existen
            if 'codigo_barras' not in df_procesado.columns or df_procesado['codigo_barras'].isna().all():
                df_procesado['codigo_barras'] = [
                    generar_codigo_barras(row.get('tecnologia'), contador + i)
                    for i, row in df_procesado.iterrows()
                ]
            
            # Agregar metadatos
            df_procesado['fecha_creacion'] = datetime.now()
            df_procesado['usuario_creador'] = 'IMPORTADOR_AUTOMATICO'
            
            # Insertar por lotes
            print(f"     üíæ Insertando en equipos_individuales...")
            
            for i in range(0, len(df_procesado), CHUNK_SIZE):
                chunk = df_procesado.iloc[i:i+CHUNK_SIZE]
                
                for _, row in chunk.iterrows():
                    try:
                        # Preparar valores
                        columnas = ', '.join(row.index)
                        placeholders = ', '.join(['?' for _ in row])
                        valores = [v if pd.notna(v) else None for v in row.values]
                        
                        cursor.execute(
                            f"INSERT INTO equipos_individuales ({columnas}) VALUES ({placeholders})",
                            valores
                        )
                        total_insertados += 1
                        contador += 1
                        
                    except Exception as e:
                        print(f"        ‚úó Error en fila: {e}")
                        continue
                
                conn.commit()
                print(f"        Insertados: {total_insertados}", end='\\r')
            
            print(f"\\n     ‚úì {total_insertados} registros insertados")
            
        except Exception as e:
            print(f"     ‚úó Error procesando archivo: {e}")
    
    # Resumen
    cursor.execute("SELECT COUNT(*) FROM equipos_individuales")
    total_final = cursor.fetchone()[0]
    
    conn.close()
    
    print("\\n" + "="*80)
    print("‚úì IMPORTACI√ìN COMPLETADA")
    print("="*80)
    print(f"\\nüìä RESUMEN:")
    print(f"   ‚Ä¢ Archivos procesados: {len(archivos)}")
    print(f"   ‚Ä¢ Registros insertados: {total_insertados:,}")
    print(f"   ‚Ä¢ Total en BD: {total_final:,}")
    print(f"\\n‚úì Base de datos actualizada: {DB_PATH}")

if __name__ == "__main__":
    main()
'''

with open(f'{final_dir}/scripts/inventario/importador_masivo_workmanager.py', 'w', encoding='utf-8') as f:
    f.write(importador_workmanager)

print("‚úì Importador masivo creado: scripts/inventario/importador_masivo_workmanager.py")
print("  - Llena equipos_individuales directamente")
print("  - Usa tus tablas existentes (sedes)")
print("  - Genera c√≥digos de barras autom√°ticos")
print("  - Sin l√≠mite de registros")
````

````python

# CREAR EXTENSIONES PARA TU M√ìDULO INVENTARIOS
extensiones_inventarios = '''"""
EXTENSIONES PARA modules/inventarios.py
Agregar estas rutas a tu m√≥dulo existente
"""

from flask import Blueprint, request, jsonify
import sqlite3

# Agregar al final de tu archivo inventarios.py

# ============================================================================
# NUEVAS RUTAS PARA B√öSQUEDA AVANZADA
# ============================================================================

@inventarios_bp.route('/api/buscar_serial_masivo', methods=['POST'])
def buscar_serial_masivo():
    """Buscar m√∫ltiples seriales a la vez"""
    try:
        data = request.get_json()
        seriales = data.get('seriales', [])
        
        if not seriales:
            return jsonify({'success': False, 'error': 'No se proporcionaron seriales'}), 400
        
        conn = sqlite3.connect('workmanager_erp.db')
        conn.row_factory = sqlite3.Row
        cursor = conn.cursor()
        
        placeholders = ','.join(['?' for _ in seriales])
        query = f"""
            SELECT * FROM equipos_individuales 
            WHERE serial IN ({placeholders})
        """
        
        cursor.execute(query, seriales)
        rows = cursor.fetchall()
        resultados = [dict(row) for row in rows]
        
        conn.close()
        
        return jsonify({
            'success': True,
            'total': len(resultados),
            'encontrados': len(resultados),
            'no_encontrados': len(seriales) - len(resultados),
            'data': resultados
        })
        
    except Exception as e:
        return jsonify({'success': False, 'error': str(e)}), 500

@inventarios_bp.route('/api/estadisticas_completas')
def estadisticas_completas():
    """Estad√≠sticas detalladas del inventario"""
    try:
        conn = sqlite3.connect('workmanager_erp.db')
        cursor = conn.cursor()
        
        # Total
        cursor.execute("SELECT COUNT(*) FROM equipos_individuales")
        total = cursor.fetchone()[0]
        
        # Por estado
        cursor.execute("""
            SELECT estado, COUNT(*) 
            FROM equipos_individuales 
            WHERE estado IS NOT NULL
            GROUP BY estado
        """)
        por_estado = {row[0]: row[1] for row in cursor.fetchall()}
        
        # Por sede
        cursor.execute("""
            SELECT s.nombre, COUNT(ei.id) as total
            FROM equipos_individuales ei
            LEFT JOIN sedes s ON ei.sede_id = s.id
            WHERE s.nombre IS NOT NULL
            GROUP BY s.nombre
            ORDER BY total DESC
        """)
        por_sede = {row[0]: row[1] for row in cursor.fetchall()}
        
        # Por tecnolog√≠a
        cursor.execute("""
            SELECT tecnologia, COUNT(*) 
            FROM equipos_individuales 
            WHERE tecnologia IS NOT NULL
            GROUP BY tecnologia
            ORDER BY COUNT(*) DESC
            LIMIT 10
        """)
        por_tecnologia = {row[0]: row[1] for row in cursor.fetchall()}
        
        # Asignados vs Disponibles
        cursor.execute("""
            SELECT 
                SUM(CASE WHEN asignado_a IS NOT NULL AND asignado_a != '' THEN 1 ELSE 0 END) as asignados,
                SUM(CASE WHEN disponible = 'SI' THEN 1 ELSE 0 END) as disponibles
            FROM equipos_individuales
        """)
        row = cursor.fetchone()
        asignados = row[0] or 0
        disponibles = row[1] or 0
        
        conn.close()
        
        return jsonify({
            'success': True,
            'data': {
                'total_equipos': total,
                'asignados': asignados,
                'disponibles': disponibles,
                'por_estado': por_estado,
                'por_sede': por_sede,
                'por_tecnologia': por_tecnologia
            }
        })
        
    except Exception as e:
        return jsonify({'success': False, 'error': str(e)}), 500

@inventarios_bp.route('/api/equipos_empleado/<cedula>')
def equipos_empleado(cedula):
    """Obtener todos los equipos asignados a un empleado"""
    try:
        conn = sqlite3.connect('workmanager_erp.db')
        conn.row_factory = sqlite3.Row
        cursor = conn.cursor()
        
        # Obtener empleado
        cursor.execute("SELECT * FROM empleados WHERE cedula = ?", (cedula,))
        empleado = cursor.fetchone()
        
        if not empleado:
            return jsonify({'success': False, 'error': 'Empleado no encontrado'}), 404
        
        # Obtener equipos
        cursor.execute("""
            SELECT * FROM equipos_individuales 
            WHERE asignado_a LIKE ?
        """, (f'%{empleado["nombre"]}%',))
        
        equipos = [dict(row) for row in cursor.fetchall()]
        
        conn.close()
        
        return jsonify({
            'success': True,
            'empleado': dict(empleado),
            'total_equipos': len(equipos),
            'equipos': equipos
        })
        
    except Exception as e:
        return jsonify({'success': False, 'error': str(e)}), 500

@inventarios_bp.route('/api/exportar_inventario_completo')
def exportar_inventario_completo():
    """Exportar inventario completo a Excel"""
    try:
        import pandas as pd
        from io import BytesIO
        from flask import send_file
        
        conn = sqlite3.connect('workmanager_erp.db')
        
        # Cargar datos
        df = pd.read_sql_query("""
            SELECT 
                ei.*,
                s.nombre as sede_nombre,
                s.ciudad as sede_ciudad
            FROM equipos_individuales ei
            LEFT JOIN sedes s ON ei.sede_id = s.id
            ORDER BY ei.id DESC
        """, conn)
        
        conn.close()
        
        # Crear Excel
        output = BytesIO()
        with pd.ExcelWriter(output, engine='openpyxl') as writer:
            df.to_excel(writer, sheet_name='Inventario Completo', index=False)
        
        output.seek(0)
        
        return send_file(
            output,
            mimetype='application/vnd.openxmlformats-officedocument.spreadsheetml.sheet',
            as_attachment=True,
            download_name=f'inventario_completo_{datetime.now().strftime("%Y%m%d_%H%M%S")}.xlsx'
        )
        
    except Exception as e:
        return jsonify({'success': False, 'error': str(e)}), 500
'''

with open(f'{final_dir}/modules/inventario_maestro/extensiones_inventarios.py', 'w', encoding='utf-8') as f:
    f.write(extensiones_inventarios)

print("‚úì Extensiones creadas: modules/inventario_maestro/extensiones_inventarios.py")
print("  - B√∫squeda masiva de seriales")
print("  - Estad√≠sticas completas")
print("  - Equipos por empleado")
print("  - Exportaci√≥n completa a Excel")
````
